```python
#5
def find_month(x):
    if " 01 " in x:
        return "Jan"
    elif " 02 " in x:
        return "Feb"
    elif " 03 " in x:
        return "March"    
    elif " 04 " in x:
        return "April"    
    elif " 05 " in x:
        return "May"    
    elif " 06 " in x:
        return "June"    
    elif " 07 " in x:
        return "July"    
    elif " 08 " in x:
        return "August"    
    elif " 09 " in x:
        return "Sep"    
    elif " 10 " in x:
        return "Oct"    
    elif " 11 " in x:
        return "Nov"    
    else:
        return "Dec"
```

```python
df["Month"]=df["Date/Time"].apply(find_month)
```

```python
gk = df.groupby('Month')
gk.mean()
#March, August, and November have the highest average wind speeds. These are the same months with the highest average
#power output. This makes sense because the power is dependent on wind speeds, so as wind speeds increase the power will
#also increase.
```

```python
#1a
df.plot.scatter('Wind Speed (m/s)','Theoretical_Power_Curve (KWh)')
plt.title('Theoretical Power Vs Wind Speed')
plt.ylabel('Theoretical Power')
#Wind Speed should be on the X axis and Theoretical Power on the y axis because the power produced by the wind turbine
#is dependant on the wind speed. Based on this scatterplot, it is apparent that as wind speeds increase from 0-10 m/s,
#the theoretical power increases. After 10 m/s, the power appears to reach a maximum and stop increasing with wind speed.
```

```python
#1b
df.plot.scatter('Wind Direction (°)','Theoretical_Power_Curve (KWh)')
plt.title('Theoretical Power Vs Wind Direction')
plt.ylabel('Theoretical Power')
#Theoretical power should be on the y axis and wind direction on the x axis because the theoretical power is the dependent
#variable in this model. Based on this scatterplot, there is no linear relationship between wind direction and theoretical
#power.
```

```python
#The two plots tell me that power in a wind turbine is highly dependant on Wind Speed, but has little to no correlation
#with wind direction. This makes sense logically because the faster the wind is going, the more the turbine will be moved, 
#which will generate more energy. A change in wind direction will not affect the power produced, becuase the turbine will
#move in either direction and produce the same amount of energy if the speed is the same.
```

```python
#2a
df.plot.scatter('Wind Speed (m/s)','LV ActivePower (kW)')
plt.title('Actual Power Vs Wind Speed')
plt.ylabel('Actual Power')
#This graph follows the shape of the theoretical power vs. Wind speed graph. The actual power increases as the wind
#speed increases up to 10 m/s and the curve flattens out and remains constant after that point.
```

```python
#2b
df.plot.scatter('Wind Direction (°)','LV ActivePower (kW)')
plt.title('Actual Power Vs Wind Direction (°)')
plt.ylabel('Actual Power')
#Again, this scatterplot follows the shape of the theoretical plot. This makes sense because Actual power is not linearly
#dependent on wind direction. It appears that the two main clusters of points are about 180 degrees apart, which makes sense
#because there are two directions that are 180 degrees apart in which the wind will have the most ability to move the turbine,
#which will generate higher amounts of power
```

```python
#3 
#The scatterplot of the theoretical power vs. wind speed and the scatterplot of the actual power vs. wind speed have very 
#similar shapes, but the theoretical power plot is a lot smoother of a curve with less outliers and points outside of the
#curve. This is because the theoretical power is a calculated prediction of how the turbine should act at different wind 
#speeds, while the actual power is actual recorded points, meaning there will be some more variation and a less ideal output.
#The plots of the theoretical power vs. wind direction and actual power vs. wind direction are extremely similar. I can not 
#tell any big differences between the two scatterplots.
```

```python
#4
fig = plt.figure(figsize=(12,9))
sns.heatmap(df.corr(),annot=True)
plt.show()
# a) Wind speed and theoretical power curve are very strongly positively correlated, with a correlation value of .94. 
#This value aligns with the scatterplot because the line of fit appears to be fairly linear.
# b) Wind direction and theoreticl power have almost no correlation, with a value of -.099. This is supported by the
#scatterplot of the same data, which does not have a linear shape.
# c) Wind speed and actual power curve have a correlation of .91, which is slightly lower than the value between wind 
#speed and theoretical power, but still shows a strong positive correlation. Based on the scatterplots, the wind speed 
#and actual power followed the same general shape as wind speed and theoretical power with a sligtly less smooth curve,
#so it makes sense that the correlation value would be only slightly lower.
# d) Wind direction and actual power have a very weak correlation, with a value of -.063. This is very close to 0, which makes
#sense because the scatterplot had no recognizable shape. 
```

```python
#5
#The theoretical power and actual power have a correlation of .95, which is a very strong positive correlation. This
#makes a lot of sense because the theoretical power curve is essentially an estimation of the actual power curve, so if 
#the estimation was calculated correctly than the two curves should be very closely correlated. 
```

```python
def wind_direction(x):
    if 0<=x<40:
        return "N"
    elif 40<=x<80:
        return "NE"
    elif 80<=x<120:
        return "E"
    elif 120<=x<160:
        return "SE"
    elif 160<=x<200:
        return "S"
    elif 200<=x<240:
        return "SW"
    elif 240<=x<280:
        return "W"
    elif 280<=x<320:
        return "NW"
    elif 320<=x<=360:
        return "N"
```

```python
df["Wind Direction"]=df["Wind Direction (°)"].apply(wind_direction)
```
